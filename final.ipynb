```python
data = data.replace('No',0)
data = data.replace('Yes',1)
data = data.replace('Female',0)
data = data.replace('Male',1)
data = data.replace('Negative',0)
data = data.replace('Positive',1)
data.head()
```

```python
y=data['class']
x=data.drop('class',axis=1)
x = np.array(x)
y = np.array(y)
```

```python
sns.heatmap(data.corr(),annot=True,linewidths=2,linecolor='black',square=True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_scaled=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_scaled,y,test_size=0.3,random_state=100)
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='linear',probability=True)
svm_model.fit(xtrain,ytrain) 
```

```python
from sklearn.metrics import accuracy_score
predictions1 = svm_model.predict(xtest)
svm_acc = accuracy_score(ytest, predictions1)
print("Accuracy svm : ",svm_acc)
```

```python
from sklearn.metrics import confusion_matrix,f1_score,precision_score,recall_score,mean_squared_error
tn, fp, fn, tp = confusion_matrix(ytest, predictions1).ravel()
```

```python
print('Precision of svm :',precision_score(ytest,predictions1)) #tp/tp+fp
```

```python
print('Recall score of svm :',recall_score(ytest,predictions1)) #tp/tp+fn
```

```python
specificity = tn / (tn + fp)
print(f"Specificity: {specificity:}")
```

```python
print('F1 score of svm :',f1_score(ytest,predictions1)) #2*p*r/(p+r)
```

```python
print('Mean sqaured error of svm :',mean_squared_error(ytest,predictions1))
```

```python
y_pred_svm=svm_model.predict(xtest)
```

```python
w=svm_model.coef_[0]
print(w)
```

```python
from sklearn.metrics import roc_curve,auc
fpr,tpr,thresholds=roc_curve(ytest,y_pred_svm)
roc_auc=auc(fpr,tpr)
plt.clf()
plt.plot(fpr, tpr, color='black', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic of SVM')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
ann_model = MLPClassifier(random_state=1, max_iter=300, solver='adam', hidden_layer_sizes=(16,)*16)
ann_model.fit(xtrain, ytrain)
```

```python
predictions = ann_model.predict(xtest)
clf_acc = accuracy_score(ytest,predictions)
print("Accuracy MLP Classifier : ", clf_acc)
```

```python
print('Precision of ANN :',precision_score(ytest,predictions))
```

```python
print('Recall of ANN: ',recall_score(ytest,predictions))
```

```python
# fusion
import numpy as np
pb_mlp = ann_model.predict_proba(xtest)
pb_svm = svm_model.predict_proba(xtest)

a= np.mean(pb_mlp)
b= np.mean(pb_svm)

w1 = a/(a+b)
w2 = b/(a+b)
w2
```

```python
from sklearn.metrics import accuracy_score
combined_proba = w1 * pb_mlp + w2 * pb_svm
final_pred=np.argmax(combined_proba,axis=1)
accuracy_score(ytest,final_pred)
```

```python
inp_data = xtest[:1,:]
pb1 = ann_model.predict_proba(inp_data)
pb2 = svm_model.predict_proba(inp_data)
pb = w1*pb1+ w2*pb2

d_class = np.argmax(pb)
if d_class==1:
  print('Positive')
else:

  print('Negative')
```
